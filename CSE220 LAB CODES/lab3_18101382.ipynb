```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def remove_compartment(head, n):
    if head is None:
        return None

    length = 0
    current = head
    while current is not None:
        length += 1
        current = current.next

    if n >= length:
        return head.next

    position = length - n

    current = head
    for _ in range(position - 1):
        current = current.next

    current.next = current.next.next

    return head


def createList(arr):
    head = Node(arr[0])
    temp = head

    for i in range(1, len(arr)):
        new_node = Node(arr[i])
        temp.next = new_node
        temp = new_node

    return head


head = createList([10, 15, 34, 41, 56, 72])
print('Original Compartment Sequence:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')

head = remove_compartment(head, 2)
print('Changed Compartment Sequence:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')

head = createList([10, 15, 34, 41, 56, 72])
print('Original Compartment Sequence:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')

head = remove_compartment(head, 7)
print('Changed Compartment Sequence:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')

head = createList([10, 15, 34, 41, 56, 72])
print('Original Compartment Sequence:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')

head = remove_compartment(head, 6)
print('Changed Compartment Sequence:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def shuffle_on_index(head):
    if head is None or head.next is None:
        return head

    # Separate the even-indexed and odd-indexed songs
    even_head = head
    odd_head = head.next
    even_current = even_head
    odd_current = odd_head

    current = odd_head.next
    is_even = False

    while current is not None:
        if is_even:
            even_current.next = current
            even_current = even_current.next
        else:
            odd_current.next = current
            odd_current = odd_current.next

        current = current.next
        is_even = not is_even


    even_current.next = odd_head
    odd_current.next = None

    return even_head


def createList(arr):
    head = Node(arr[0])
    temp = head

    for i in range(1, len(arr)):
        new_node = Node(arr[i])
        temp.next = new_node
        temp = new_node

    return head


head = createList(['S', 'E', 'N', 'P', 'A', 'I'])
print('Original Play List:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')

new_head = shuffle_on_index(head)
print('New Play List:', end=' ')
current = new_head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')

head = createList(['N', 'I', 'S', 'H', 'I', 'N', 'O', 'Y', 'A'])
print('Original Play List:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')

new_head = shuffle_on_index(head)
print('New Play List:', end=' ')
current = new_head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def find_next_clue(head):
    if head is None:
        return

    current = head

    while current is not None:
        next_greater = find_next_greater(current)
        if next_greater is None:
            print("No next element greater than", current.data)
        else:
            print(next_greater.data, "is greater than", current.data)

        current = current.next


def find_next_greater(node):
    current = node.next

    while current is not None:
        if current.data > node.data:
            return current
        current = current.next

    return None


def createList(arr):
    head = Node(arr[0])
    temp = head

    for i in range(1, len(arr)):
        new_node = Node(arr[i])
        temp.next = new_node
        temp = new_node

    return head



head = createList([7, 85, 54, 16, 11, 30])
print('Given Clue List:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')
find_next_clue(head)
print()

head = createList([20, 13, 33, 12])
print('Given Clue List:', end=' ')
current = head
while current:
    print(current.data, end='→')
    current = current.next
print('\b\b ')
find_next_clue(head)
print()
```
