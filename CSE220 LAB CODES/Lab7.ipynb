```python
class Node:
  def __init__(self, elem, next = None):
    self.elem, self.next = elem, next

def create_linked_list(arr):
  head = Node(arr[0])
  tail = head
  for i in arr[1:]:
    new_node = Node(i)
    tail.next = new_node
    tail = new_node
  return head

def count(head):
  count = 0
  while head != None:
    count += 1
    head = head.next
  return count

def print_linked_list(head):
  while head != None:
    print(head.elem, end = '-> ')
    head = head.next
  print('None')
  print()
```

```python
def nerdy_run(path, k):

    hash_set = {}

    for i in range(len(path)):
        num = path[i]


        for j in range(max(0, i - k), i):
            if path[j] == num:
                return num


        hash_set[num] = i


        if i >= k:
            del hash_set[path[i - k]]

    return None

```

```python
def main():

  path = [6,7,8,9,5,9]
  k = 3
  result = nerdy_run(path,k)
  unittest.output_test(result, 9)
  print(f'Duplicate value within range {k} is: {result}') #Duplicate value within range 3 is: 9

  print('===================================')
  path = [6,7,8,9,5,6]
  k = 4
  result = nerdy_run(path,k)
  unittest.output_test(result, None)
  print(f'Duplicate value within range {k} is: {result}') #Duplicate value within range 4 is: None

  print('===================================')
  path = [0.21,1.21,4.67,0.21,0.45,1.9]
  k = 7
  result = nerdy_run(path,k)
  unittest.output_test(result, 0.21)
  print(f'Duplicate value within range {k} is: {result}') #Duplicate value within range 7 is: 0.21


if __name__ == "__main__":
  main()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Hashtable:
    def __init__(self, size):
        self.ht = [None] * size
        self.size = size

    def insert(self, s):
        if self.search_hashtable(s) == 'Found':
            print(s, 'Already Inserted. Cannot reinsert.')
            print('===============================')
            return
        key = self.__hash_function(s)
        node = Node(s)
        if self.ht[key] is None:
            self.ht[key] = node
        else:
            node.next = self.ht[key]
            self.ht[key] = node

    def create_from_array(self, arr):
        for i in arr:
            self.insert(i)

    def print_hashtable(self):
        idx = 0
        for i in self.ht:
            print(idx, ':', end=' ')
            self.print_linked_list(i)
            idx += 1

    def print_linked_list(self, node):
        while node:
            print(node.data, end=" -> ")
            node = node.next
        print("None")

    def __hash_function(self, s):
        total = 0
        for char in s:
            total += ord(char)
        return total % self.size

    def search_hashtable(self, s):
        key = self.__hash_function(s)
        node = self.ht[key]
        while node:
            if node.data == s:
                return 'Found'
            node = node.next
        return 'Not Found'
```
